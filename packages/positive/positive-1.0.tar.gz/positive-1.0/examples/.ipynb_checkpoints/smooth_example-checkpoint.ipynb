## Generate Test Data

```python
t = pi*linspace( -1,1,2e2 )
noise = array( [ random.uniform( -1,1 ) for k in t ] )
y =  noise*0.1*(1+1*t) + cos(2*t) * ( sin( 0.5*t ) )
plot( t, y )
xlim(lim(t))
xlabel('$x$')
ylabel('$y$')
```

## Apply smooth class to data

```python
# Use auto width determination
print 'USING Savlo filter'
foo = smooth( y )
foo.plot()
show()
# Use input width 
print 'USING Rolling Average'
bar = smooth( y, 60, auto_method='average', method='auto' )
bar.plot()
```
